## Importing required Libraries

```python
import pandas as pd
import numpy as np
from google.colab import files
import io
```

## Importing csv file

```python
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['FoodBalanceSheets_E_Africa_NOFLAG.csv']), encoding="latin-1")
df.head()
```

## Question 1 - Number and Percentage of missing data in 2014 to 3 decimal places

```python
print("Total Number -",df['Y2014'].isnull().sum())

print("Percentage -",round(((df['Y2014'].isnull().sum())*100)/len(df),3))
```

## Question 2 - Perform groupby on Element. Total sum of Processing in 2017

```python
df.groupby('Element')['Y2017'].sum()
```

## Question 5 - Highest Sum Area in 2017

```python
df.groupby('Area')['Y2017'].sum()
```

```python
df.groupby('Area')['Y2017'].sum().max()
```

Here we can see the max value is 1483268.23 which belongs to Nigeria<br><br>
## Question 6 - Area with 7th lowest Sum in 2017

```python
sorted(df.groupby('Area')['Y2017'].sum())
```

As we can see the 7th lowest value is 19102.77<br>
This value belongs to Guinea Bissau<br><br>
